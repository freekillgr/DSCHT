## Preprocessing

3 mandatory steps:
- **handle missing values**
- **encode categorical variables**
- feature scaling

Optional steps:
- **cleaning**
- feature selection/extraction
- resampling
- etc.

### Cleaning

Remove whitespace from strings

````python
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
````

### Missing values

````python
df.isna().sum()
````

Replace `'?'` with `np.nan`

````python
df = df.applymap(lambda s: np.nan if s == '?' else s)
````

````python
df.isna().sum()
````

Fill missing values with most common element in respective column.

````python
df = df.fillna(df.mode().iloc[0])

# df = df.dropna()  # drop all rows that have a missing value
````

````python
df.isna().sum()
````
