## Preprocessing

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns='Concrete compressive strength')
y = df['Concrete compressive strength']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_estimators': [10, 50, 100, 200, 1000],
    'criterion': ['squared_error', 'absolute_error'],
    'max_depth': [None, 7, 5, 3, 2],
    'min_samples_split': [2, 5, 8]
}

rf = RandomForestRegressor()


gs = GridSearchCV(rf, param_grid, scoring='neg_mean_squared_error', verbose=10, n_jobs=-1)

gs.fit(x_train, y_train)
```

```python
gs.cv_results_['mean_test_score']
```

```python
gs.cv_results_.keys()
```

```python
n_estimators_values = gs.cv_results_['param_n_estimators']
min_samples_values = gs.cv_results_['param_min_samples_split']
criterion_values = gs.cv_results_['param_criterion']
max_depth_values = gs.cv_results_['param_max_depth']

mean_results = gs.cv_results_['mean_test_score']

results = pd.DataFrame({
    'n_estimators': n_estimators_values,
    'min_samples_split': min_samples_values,
    'criterion': criterion_values,
    'max_depth': max_depth_values,
    'score': mean_results})

results = results.fillna({'max_depth': 10}) # fill None max_depth with 10
```

```python
sns.boxplot(x='n_estimators', y='score', data=results)
```

```python
sns.boxplot(x='min_samples_split', y='score', data=results)
```

```python
sns.boxplot(x='max_depth', y='score', data=results)
```

```python
sns.boxplot(x='criterion', y='score', data=results)
```

```python
results2 = results[results['max_depth'] == 10]
```

```python
sns.boxplot(x='n_estimators', y='score', data=results2)
```

```python
sns.boxplot(x='n_estimators', y='score', hue='criterion', data=results2)
```

```python
sns.boxplot(x='n_estimators', y='score', hue='min_samples_split', data=results2)
```

```python
results3 = results2[results2['min_samples_split'] == 2]
```

```python
results3.sort_values('score', ascending=False)
```

Best results are certainly with **max_depth=10** (i.e. None) and **min_samples_split=2**. Both are the default values!

Regarding the other two seems to be with **absolute error** and with **n_estimators** on the high end (above 200).

We'll now do a more **targeted** grid search

```python
param_grid = {
    'n_estimators': [200, 500, 1000, 1500]
}

rf = RandomForestRegressor(min_samples_split=2, max_depth=None, criterion='absolute_error')

gs = GridSearchCV(rf, param_grid, scoring='neg_mean_squared_error', verbose=10, n_jobs=-1)

gs.fit(x_train, y_train)
```
